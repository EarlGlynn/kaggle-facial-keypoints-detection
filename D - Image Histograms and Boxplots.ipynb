## plotTrainImage function

```r
plotTrainImage <- function(TRAIN)
{
  showImage(trainImage(TRAIN))
  showFeatures(dTrain, TRAIN)
  text(2,95, paste("TRAIN", TRAIN), adj=0, col="blue")

  d <- trainImage(TRAIN)

  boxplot(d, horizontal=TRUE, xaxt="n", yaxt="n")
  boxplot(t(d), xaxt="n", yaxt="n")

  x <- as.numeric(d)
  hist(x, xaxt="n", yaxt="n", breaks=255, main="")
}       
```
