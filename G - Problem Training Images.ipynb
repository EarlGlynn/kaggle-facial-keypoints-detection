```r
showImageWithFeatures <- function(TRAIN, issue)
{
  showImage(trainImage(TRAIN))
  showFeatures(dTrain, TRAIN)
  title(paste("Train", TRAIN, '-', issue), col.main="blue")
}   
```

```r
options(repr.plot.width=8, repr.plot.height=8)
par(mfrow=c(3,3), mar=c(1,1,1,1))

showImageWithFeatures(1688, "missing eye centers")
showImageWithFeatures(1748, "bad placement")
showImageWithFeatures(1862, "one side of face")
showImageWithFeatures(1878, "bad placement")
showImageWithFeatures(1908, "bad placement")
showImageWithFeatures(2091, "missing eye feature; other")
showImageWithFeatures(2138, "sunglasses; ice cream")
showImageWithFeatures(2154, "sunglasses")
showImageWithFeatures(2187, "sunglasses")
showImageWithFeatures(2195, "multiple people")
showImageWithFeatures(2200, "bad placement")
showImageWithFeatures(4264, "multiple people")
showImageWithFeatures(4491, "multiple people")
showImageWithFeatures(4829, "multiple people")
showImageWithFeatures(6406, "very small head shot")
showImageWithFeatures(6493, "multiple people")
showImageWithFeatures(6494, "multiple people")
```
